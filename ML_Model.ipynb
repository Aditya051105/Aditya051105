```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2529) #for traaining the model according to the dataset
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr=LinearRegression()# Call the linear regression model .
lr.fit(x_train,y_train)# for fitting the model with train data
```

```python
lr.score(x_test,y_test)*100 # for testing the values of dataset
```

```python
Year = float(input("Enter your Experience Years : "))
print(lr.predict([[Year]]))
```
